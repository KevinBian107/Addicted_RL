### Compariosn to random walk

```python
num_trials = 100
num_steps = 1000
num_states = 10

avg_durations = random_walk(num_trials, num_steps, num_states)

plt.figure(figsize=(8, 5))
plt.bar(range(num_states), avg_durations)
plt.title('Average Duration in Each State During Random Walk')
plt.xlabel('State')
plt.ylabel('Average Duration')
plt.xticks(range(num_states))
plt.show()
```

```python
alpha = 0.1
gamma = 0.9
epsilon = 0.1
num_trials = 10
num_states = 10
num_actions = 3
initial_dopamine_surge = 1
dopamine_decay_rate = 0.99
reward_states = [5, 9]
drug_reward = 1
addicted = True

_, Q_across_trials = addicted_q_learning(alpha, gamma, epsilon, num_trials,
                                           num_states, num_actions, initial_dopamine_surge,
                                           dopamine_decay_rate, reward_states, drug_reward, addicted)

avg_durations = simulate_state_durations(Q_across_trials[-1], num_trials, num_states)

plt.figure(figsize=(8, 5))
plt.bar(range(num_states), avg_durations)
plt.title('Average Duration in Each State (Q-Learning Agent)')
plt.xlabel('State')
plt.ylabel('Average Duration')
plt.xticks(range(num_states))
plt.show()
```
